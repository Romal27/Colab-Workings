```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin,cos,pi
```

```python
u0 = 40 #projectile velocity
g = 9.81 #accelaration/gravity

```

```python
theta1 = 45 # angle of the projectile
theta2 = 60
theta3 = 30
ux1 = u0*cos(theta1* pi/180)
uy1 = u0 * sin(theta1 * pi/180)

ux2 = u0*cos(theta2* pi/180)
uy2 = u0 * sin(theta2 * pi/180)

ux3 = u0*cos(theta3* pi/180)
uy3 = u0 * sin(theta3 * pi/180)
```

```python
t_total_1 = 2*uy1/g #total time of flight
t_total_2 = 2*uy2/g
t_total_3 = 2*uy3/g

t1 = np.linspace(0,t_total_1,100)
t2 = np.linspace(0,t_total_2,100)
t3 = np.linspace(0,t_total_3,100)

sx1 = ux1*t1
sy1 = uy1*t1 - 0.5*g*t1**2

sx2 = ux2*t2
sy2 = uy2*t2 - 0.5*g*t2**2

sx3 = ux3*t3
sy3 = uy3*t3 - 0.5*g*t3**2
```

```python
plt.figure(figsize=(10,6))
plt.plot(sx1,sy1,label = r'$\theta={}$' .format(theta1))
plt.plot(sx2,sy2,label = r'$\theta={}$' .format(theta2))
plt.plot(sx3,sy3,label = r'$\theta={}$' .format(theta3))
plt.legend(fontsize=12)
plt.title("Projectile Motion of a Ball", fontsize=12)
plt.xlabel("Horizontal Distance (m)", fontsize=12)
plt.ylabel("Vertical Distance (m)", fontsize=12)
plt.grid(True)

plt.show()
```
