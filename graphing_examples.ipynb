```python
# SECTION 1: Basic Plotting (Implicit Figure and Axes)
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)  # Matplotlib creates figure and axes implicitly
plt.title("Basic Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()

# Run this cell. You'll see a basic sine wave plot.
```

```python
# SECTION 2: Multiple Plots on the SAME Figure (Overlapping)
plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)")
plt.title("Overlapping Sine and Cosine")
plt.legend()  # Show the legend
plt.show()

# Run this cell. The cosine wave will be plotted on top of the sine wave from the previous cell.
```

```python
# SECTION 3: Multiple Figures (Separate Plots)
plt.figure()  # Create a new figure
plt.plot(x, np.sin(x))
plt.title("Separate Sine Wave")
plt.show()

# Run this cell. A new plot window with only the sine wave will appear.
```

```python
plt.figure()  # Create another new figure
plt.plot(x, np.cos(x))
plt.title("Separate Cosine Wave")
plt.show()

# Run this cell. Another new plot window with only the cosine wave will appear.
```

```python
# SECTION 4: Controlling Figure Size and DPI
plt.figure(figsize=(10, 5), dpi=150)  # Width=10 inches, Height=5 inches, 150 DPI
plt.plot(x, np.tan(x))
plt.title("Tangent Wave (Larger Figure)")
plt.ylim(-10, 10) #Limit y-axis to avoid extreme values
plt.show()

# Run this cell. You'll see a larger, higher-resolution tangent plot. Experiment with figsize and dpi values.
```

```python
# SECTION 5: Subplots (Multiple Plots on ONE Figure with Sub-Axes)
plt.figure(figsize=(10, 6))  # Figure for subplots
plt.subplot(2, 1, 1)  # 2 rows, 1 column, first subplot (top)
plt.plot(x, np.exp(-x))
plt.title("Exponential Decay")

plt.subplot(2, 1, 2)  # 2 rows, 1 column, second subplot (bottom)
plt.plot(x, x**2)
plt.title("x squared")
plt.tight_layout() #Adjust subplot params so that subplots are nicely fit in the figure.
plt.show()

# Run this cell. You'll see two plots (exponential decay and x squared) stacked vertically in the same figure.
```

```python
# SECTION 6: Object-Oriented Interface (More Control)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))  # Create a figure and 2x2 axes grid

axes[0, 0].plot(x, np.sqrt(x)) # Top-left subplot
axes[0, 0].set_title("Square Root")

axes[0, 1].plot(x, np.log(x)) # Top-right subplot
axes[0, 1].set_title("Natural Logarithm")

axes[1, 0].plot(x, x**3) # Bottom-left subplot
axes[1, 0].set_title("x cubed")

axes[1, 1].plot(x, 1/x) # Bottom-right subplot
axes[1, 1].set_title("1/x")
fig.suptitle("Multiple plots using object oriented interface") #Add a title to the whole figure
plt.tight_layout() #Adjust subplot params so that subplots are nicely fit in the figure.
plt.show()

# Run this cell. You'll see four plots arranged in a 2x2 grid within the same figure.
```

```python
# SECTION 7: Saving Figures to Files
plt.plot(x, np.sin(x))
plt.savefig("sine_wave.png")  # Save as PNG
plt.savefig("sine_wave.pdf")  # Save as PDF
plt.savefig("sine_wave.jpg", dpi = 300) # Save as JPEG with 300 dpi
plt.show() #If you don't put this line the plot will not show but the files will be saved

print("Figures saved to sine_wave.png, sine_wave.pdf, and sine_wave.jpg")

# Run this cell. The plot will be saved as files. In Google Colab, you can find the files in the "Files" tab on the left.
```
